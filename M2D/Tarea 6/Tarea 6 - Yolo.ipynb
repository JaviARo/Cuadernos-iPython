## Script de iniciación

```python
from ultralytics import YOLO
# Instancia del modelo YOLO con los pesos especificados
pesos_yolo = "yolov8n.pt"
model = YOLO(pesos_yolo)
try:
	# Realiza la detección de objetos en el flujo de video de la cámara del computador
	results = model(source=0, show=True, conf=0.3, save=True)
except Exception as e:
	print("Ocurrió un error al realizar la detección de objetos:", e)
```

## Script de tracking

```python
from ultralytics import YOLO 
# Configure the tracking parameters and run the tracker 
model = YOLO('yolov8n.pt') 
results = model.track(source="0", conf=0.3, iou=0.5, show=True)
# classes = [] array de clases
```

## Script entrenado el modelo

```python
from ultralytics import YOLO 
# Configure the tracking parameters and run the tracker 
model = YOLO('best.pt') 
results = model.track(source="0", conf=0.2, iou=0.5, show=True)
```

```python
from ultralytics import YOLO 
# Cargar el modelo YOLO preentrenado 
model = YOLO("best.pt") 
# Realizar la predicción en una imagen 
# img_name = "19_691_jpg.rf.1f2d00112f6d91125cc0d178ab4eb3e2.jpg"
# img_name = "22_99_jpg.rf.1094da02e3b0bd9cc72930f20ccc6db7.jpg"
img_name = "20_4_jpg.rf.ddb786f4f8582c8a0843fbe47352526b.jpg"
results = model(f"./food_dataset/valid/images/{img_name}", conf=0.3,iou=0.5) 
# Iterar sobre cada resultado en la lista 
for result in results: 
	# Mostrar las cajas
	print(result.boxes)
	result.show()
	result.save(filename= f"predict/boxes_{img_name}")
```
